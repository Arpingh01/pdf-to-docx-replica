```python
from docx import Document
from docx.shared import Pt
from docx.enum.text import WD_ALIGN_PARAGRAPH
from docx.enum.table import WD_TABLE_ALIGNMENT, WD_CELL_VERTICAL_ALIGNMENT

# Create document
doc = Document()

# ========== HEADER ==========
def center_bold(text, size=12):
    p = doc.add_paragraph(text)
    p.alignment = WD_ALIGN_PARAGRAPH.CENTER
    run = p.runs[0]
    run.bold = True
    run.font.size = Pt(size)

center_bold("FORM ‘A’", 14)
center_bold("MEDIATION APPLICATION FORM", 12)

p = doc.add_paragraph("[REFER RULE 3(1)]")
p.alignment = WD_ALIGN_PARAGRAPH.CENTER

doc.add_paragraph("Mumbai District Legal Services Authority").alignment = WD_ALIGN_PARAGRAPH.CENTER
doc.add_paragraph("City Civil Court, Mumbai\n").alignment = WD_ALIGN_PARAGRAPH.CENTER


# ========== MAIN TABLE ==========
table = doc.add_table(rows=1, cols=3)
table.alignment = WD_TABLE_ALIGNMENT.CENTER
table.style = "Table Grid"

# Helper to add a full-width row
def full_row(text, bold=True):
    row = table.add_row().cells
    row[0].merge(row[1]).merge(row[2])
    p = row[0].paragraphs[0]
    p.text = text
    if bold:
        p.runs[0].bold = True

# DETAILS OF PARTIES
full_row("DETAILS OF PARTIES:")

# ---------- APPLICANT ----------
row = table.add_row().cells
row[0].text = "1"
row[1].text = "Name of Applicant"
row[2].text = "{{client_name}}"

full_row("Address and contact details of Applicant")

row = table.add_row().cells
row[0].text = "1"
row[1].text = "Address"
row[2].text = (
    "REGISTERED ADDRESS:\n"
    "{{branch_address}}\n\n"
    "CORRESPONDENCE BRANCH ADDRESS:\n"
    "{{branch_address}}"
)

row = table.add_row().cells
row[0].text = ""
row[1].text = "Telephone No."
row[2].text = "{{mobile}}"

row = table.add_row().cells
row[0].text = ""
row[1].text = "Mobile No."
row[2].text = ""

row = table.add_row().cells
row[0].text = ""
row[1].text = "Email ID"
row[2].text = "info@kslegal.co.in"

# ---------- OPPOSITE PARTY ----------
row = table.add_row().cells
row[0].merge(row[1]).merge(row[2])
row[0].paragraphs[0].text = "2   Name, Address and Contact details of Opposite Party:"
row[0].paragraphs[0].runs[0].bold = True

full_row("Address and contact details of Defendant/s")

row = table.add_row().cells
row[0].text = ""
row[1].text = "Name"
row[2].text = "{{customer_name}}"

row = table.add_row().cells
row[0].text = ""
row[1].text = "Address"
row[2].text = (
    "REGISTERED ADDRESS:\n"
    "{% if address1 and address1 != '' %}{{address1}}{% else %}____________{% endif %}\n\n"
    "CORRESPONDENCE ADDRESS:\n"
    "{% if address1 and address1 != '' %}{{address1}}{% else %}____________{% endif %}"
)

row = table.add_row().cells
row[0].text = ""
row[1].text = "Telephone No."
row[2].text = ""

row = table.add_row().cells
row[0].text = ""
row[1].text = "Mobile No."
row[2].text = ""

row = table.add_row().cells
row[0].text = ""
row[1].text = "Email ID"
row[2].text = ""

# ---------- DETAILS OF DISPUTE ----------
full_row("DETAILS OF DISPUTE:")

row = table.add_row().cells
row[0].merge(row[1]).merge(row[2])
row[0].paragraphs[0].text = "THE COMM. COURTS (PRE-INSTITUTION …… SETTLEMENT) RULES, 2018"
row[0].paragraphs[0].runs[0].bold = True

row = table.add_row().cells
row[0].merge(row[1]).merge(row[2])
row[0].paragraphs[0].text = (
    "Nature of disputes as per section 2(1)(c) of the Commercial Courts Act, 2015 (4 of 2016):"
)

# Save document
doc.save("Mediation_Application_Form_FINAL.docx")

```
