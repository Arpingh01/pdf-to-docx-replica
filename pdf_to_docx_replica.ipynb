# PDF to DOCX Replica

This notebook converts a PDF file into a DOCX file using Python.

```python
import fitz  # PyMuPDF
from docx import Document

def pdf_to_docx(pdf_path, docx_path):
    doc = Document()
    pdf = fitz.open(pdf_path)

    for page in pdf:
        text = page.get_text()
        doc.add_paragraph(text)

    doc.save(docx_path)

print("Ready to convert PDF to DOCX")
```
